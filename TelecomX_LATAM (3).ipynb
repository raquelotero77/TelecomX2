```python
from pandas import json_normalize
import pandas as pd
import json

data = pd.read_json('https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
data
df = json_normalize(data['customerID'], sep='')
df
```
